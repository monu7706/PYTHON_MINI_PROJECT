#### Handling nan values ideas

```python
# - Text columns (product, brand, description)
# - Bahut kam NaN hain to .dropna() ya fillna("Unknown")
# - - Numeric columns (sale_price, market_price, rating)
# - Mean/median se fill karo.
```

```python
# numeric
for col in ['sale_price', 'market_price', 'rating','discount_amount','discount_percent']:
    df[col] = df[col].fillna(df[col].mean())
```

```python
# categorical / text
for col in ['brand', 'category', 'sub_category', 'type', 'description','product']:
    df[col] = df[col].fillna('Unknown')
```

```python
df.isnull().sum()
```

### Step 8: Find out the outliers from the dataset according to
### the columns

#### 1. To see visuals by Boxplot

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x=df['sale_price'])
plt.title("Sale Price Distribution with Outliers")
plt.show()
```

#### 2. - “We use IQR method to detect outliers.”


```python
Q1 = df['sale_price'].quantile(0.25)
Q3 = df['sale_price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Flag outliers
outliers = df[(df['sale_price'] < lower_bound) | (df['sale_price'] > upper_bound)]

outliers.shape
```

#### - “We replace extreme values with mean to reduce skewness.”

```python
mean_price = df['sale_price'].mean()
df.loc[(df['sale_price'] < lower_bound) | (df['sale_price'] > upper_bound), 'sale_price'] = mean_price
```

### Step 9: Create Plots or visualizations.

#### 9.1 Category distribution (countplot)

```python
plt.figure(figsize=(12, 6))
top_cats = df['category'].value_counts().head(10)

sns.barplot(x=top_cats.index, y=top_cats.values)
plt.xticks(rotation=45, ha='right')
plt.title("Top 10 Product Categories by Count")
plt.xlabel("Category")
plt.ylabel("Number of Products")
plt.tight_layout()
plt.show()
```

#### 9.2 Average rating by category

```python
category_rating = df.groupby('category')['rating'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=category_rating.index, y=category_rating.values)
plt.xticks(rotation=45, ha='right')
plt.title("Top 10 Categories by Average Rating")
plt.xlabel("Category")
plt.ylabel("Average Rating")
plt.tight_layout()
plt.show()
```
