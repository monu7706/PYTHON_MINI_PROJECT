### To check missing values

```python
# Missing values count column-wise
df.isnull().sum()
```

```python
# Example: fill numeric missing with median
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())

# Example: fill categorical missing with mode
categorical_cols = df.select_dtypes(include=['object']).columns

for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.isnull().sum()
```

###  – Check Outliers 

```python
# Boxplot for annual premium
plt.figure(figsize=(8, 4))
sns.boxplot(x=df['Annual_Premium'])
plt.title("Boxplot of Annual Premium")
plt.show()
```

```python
# Remove extreme outliers in annual_premium
q99 = df['Annual_Premium'].quantile(0.99)
df = df[df['Annual_Premium'] <= q99]
```

### 2. Age distribution + impact on claims

```python
plt.figure(figsize=(8, 4))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title("Age Distribution of Customers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# Create age bins
df['Age'] = pd.cut(df['Age'], bins=[18, 25, 35, 45, 55, 65, 80],
                       labels=['18-25', '26-35', '36-45', '46-55', '56-65', '66-80'])

age_claim = df.groupby('Age')['Response'].mean().reset_index()

plt.figure(figsize=(8, 4))

sns.barplot(data=age_claim, x='Age', y='Response')
plt.title("Average Claim Rate by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Claim Rate")
plt.show()
```

#### “Claim rate is higher in age group 36–45 compared to younger groups”

### 8. Region-wise analysis

```python
region_claim = df.groupby('Region_Code')['Response'].mean().reset_index()

# Top 10 regions by claim rate
top_regions = region_claim.sort_values(by='Response', ascending=False).head(10)

plt.figure(figsize=(10, 5))
sns.barplot(data=top_regions, x='Region_Code', y='Response')
plt.title("Top 10 Regions by Claim Rate")
plt.xlabel("Region Code")
plt.ylabel("Claim Rate")
plt.xticks(rotation=45)
plt.show()
```

### 9. Policy analysis

```python
policy_claim = df.groupby('Policy_Sales_Channel')['Response'].mean().reset_index()

# Top 10 channels
top_policy = policy_claim.sort_values(by='Response', ascending=False).head(10)

plt.figure(figsize=(10, 5))
sns.barplot(data=top_policy, x='Policy_Sales_Channel', y='Response')
plt.title("Claim Rate by Policy Sales Channel (Top 10)")
plt.xlabel("Policy Sales Channel")
plt.ylabel("Claim Rate")
plt.xticks(rotation=45)
plt.show()
```

### 10. Customer loyalty (number of policies vs claim)

```python

df['Annual_Premium'] = df['Policy_Sales_Channel'].apply(lambda x: x if x <= 5 else 5)

policies_claim = df.groupby('Annual_Premium')['Response'].mean().reset_index()

plt.figure(figsize=(6, 4))
sns.barplot(data=policies_claim, x='Annual_Premium', y='Response')
plt.title("Claim Rate by Number of Policies (Capped at 5+)")
plt.xlabel("Number of Policies")
plt.ylabel("Claim Rate")
plt.show()
```

## Key Findings

### 1. Age Groups with Higher Claims

##### Certain age groups show a higher claim rate compared to others. Middle‑aged customers (for example, 30–45 years) often have slightly higher claim tendencies than very young or older customers.

### 2. Gender differences

##### There is a small difference in claim rates in between male and female customers. However the difference is not always very large and depends on the datasets.

### 3. Impact of Vehicle Age and Vehicle Damage

##### - Older vehicles generally show a higher chance of claims.
##### - Customers whose vehicles were previously damaged (“Yes”) have a significantly higher claim rate compared to those with no past damage.


### 4. Premium Amount vs Claim Behavior

##### High‑premium customers do not always claim more. In many cases, customers with medium‑range premiums show more claim activity. Premium alone is not a strong indicator of claim likelihood.

### 5. Risky Regions / Policy Channels

##### Some region or Policy Sales Channels shows higher claim rates than others. these areas may requires closer monitoring or sticker verification. 

### 6. Customer Loyalty (Number of Policies)

##### Customers with more policies (loyal customers) may show different claim patterns. In some cases, customers with multiple policies tend to claim slightly more, possibly due to higher engagement with the insurer.

## Business Insights

### 1. Premium Adjustment Opportunities

##### Premiums can be adjusted for:
- Customers with older vehicles
- Customers with a history of vehicle damage
- Regions with consistently high claim rates
These groups may require higher premiums or additional risk checks.


### 2. Extra Verification or Safery Campaigns

##### Extra verification or awareness campaigns can be targeted toward:
- High‑risk regions
- Customers with damaged vehicles
- Age groups with higher claim frequency
This can help reduce fraudulent or unnecessary claims.


### 3. Stricter Underwriting Requirements

##### Stricter underwriting rules may be needed for:
- Specific policy sales channels with high claim rates
- Regions showing repeated high‑risk patterns
- Customers with multiple past claims or vehicle damage history
